```python
import pandas as pd
import pickle
import xgboost as xgb
```

```python
with open('/home/rjsietsma/PycharmProjects/capice_updatr/datafiles/Feature_importances/xgb_booster.pickle.dat', 'rb') as model_file:
    model = pickle.load(model_file)
```

```python
importances = model.get_score(importance_type='gain')
```

```python
x = pd.DataFrame(importances.values(), index=importances.keys())
x.columns=['gain']
```

```python
x.sort_values(by='gain', ascending=False)[:20]
```

```python
VEP = ['Length', 'ConsDetail',
       'oAA', 'nAA',
       'cDNApos', 'relcDNApos', 
       'CDSpos', 'relCDSpos', 
       'protPos', 'relProtPos',
       'Domain', 'motifEHIPos',
       'motifEScoreChng', 'motifEName',
       'motifECount', 'PolyPhenCat', 'PolyPhenVal',
       'GC', 'CpG', 'SIFTcat', 'SIFTval', 'Type', 'Ref', 'Alt'
      ]
```

```python
for ind in VEP:
    curr = x[x.index.str.startswith(ind)]
    if curr.shape[0] > 0:
        x.drop(index=curr.index, inplace=True)
```

```python
x.sort_values(by='gain', ascending=False)[:20]
```
