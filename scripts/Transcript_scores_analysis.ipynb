```python
from pathlib import Path
import os
import pandas as pd
```

```python
root = Path(os.path.dirname(os.path.realpath('__file__'))).parent
file_loc = os.path.join(root, 'datafiles', 'Transcript_scores_analysis', 'VEP_converted_output_capice.tsv.gz')
```

```python
data = pd.read_csv(file_loc, sep='\t')
data
```

```python
duplicates = data[data.duplicated(subset=['chr_pos_ref_alt', 'GeneName'], keep=False)]
duplicates
```

```python
for gene in duplicates['GeneName'].unique():
    print(f'For gene: {gene}')
    counts = duplicates.loc[duplicates[duplicates['GeneName'] == gene].index, :]
    counts.sort_values(by='probabilities', ascending=False, inplace=True)
    print(f'{len(counts["probabilities"].unique())}/{len(counts["probabilities"])} unique scores were found.')
    for score in counts['probabilities'].unique():
        subset = counts.loc[counts[counts['probabilities'] == score].index, :]
        n = subset.shape[0]
        features = list(subset['FeatureID'])
        if n > 1:
            print(f'Score {score} is found in {len(features)} transcripts: {", ".join(features)}')
        else:
            print(f'Score {score} is only found in transcript: {features[0]}')
    print('\n')
```
